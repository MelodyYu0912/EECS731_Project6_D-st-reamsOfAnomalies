## 1.2 Verify if the data is normal distributed
- Use kstest and normaltest to verify the observation above.

```python
from scipy.stats import kstest
from scipy.stats import normaltest

print(kstest(data["value"], 'norm'))
print(normaltest(data["value"], axis = None))
```

#### Conclusion
- Both p values are really low, which indicates that the dataset is not normally distributed. Low p values could lead to inaccuracy when applying some anomaly detection algorithms.

```python
data_60 = data.loc[data["value"] >= 60]
sns.distplot(data_60["value"]);
```

```python
print(kstest(data_60["value"], 'norm'))
print(normaltest(data_60["value"], axis = None))
```

#### Notes:

## 2. Anomaly Detection

### 2.1  3σ Detection

```python
def three_sigma(Ser1):
    rule = (Ser1.mean()-3*Ser1.std()>Ser1) | (Ser1.mean()+3*Ser1.std()< Ser1)
    index = np.arange(Ser1.shape[0])[rule]
    outrange = Ser1.iloc[index]
    return outrange

data_anomaly = three_sigma(data['value'])
print("Count:", data_anomaly.count())
print("Anomalies:")
data_anomaly
```

```python
data60_anomaly = three_sigma(data_60['value'])
print("Count:", data60_anomaly.count())
print("Anomalies:")
data60_anomaly
```

#### Conclusion:
- The 3-sigma algorithm successfully detects 12 outliers on the orginal dataset, including 11 outliers that is below 60 and were dropped for the modified dataset. The 2 result dataset matched with each other.

### 2.2 Isolation Forest

```python
data['Weekday'] = data.index.map(lambda s : s.weekday())
data['Time'] = data.index - data.index.map(lambda s : s.floor(freq ='D'))
data['Time'] = data.Time.map(lambda s : s.value/60/60/1000000000)
data.head()
```

```python
from sklearn.ensemble import IsolationForest

rng = np.random.RandomState(42)
ilf = IsolationForest(n_estimators = 100,
                      max_samples = 256,
                      behaviour = "new",
                      random_state = rng,
    )

X_cols = ["value", "Weekday", "Time"]
ilf.fit(data[X_cols])
pred = ilf.predict(data[X_cols])
 
data_pred = data
data_pred["pred"] = pred

data_pred.loc[data_pred["pred"] == -1]
```

#### Conclusion
- By applying the naive Bayes Rules, the data that appears less frequently will be considered as anomalies, which works fine since it is a one dimensional data set. 
- Looking into the anomalies given by applying the naive Bayes Rule, data such as 109 and 43 are outliers, which make sense because the firs one is overspeed and the second is underspeed.
- After applying the naive Bayes Rule, I applied isolation forest. 
- For this model I also passed in time as a feature ,since time may affect hoe fast people drive, e.g people may drive faster during the midnight when there are less cars on the street. I converted the timestamp to its conrresponding numerical value by appylying math equations. 
- Isolation forest also gives anomalies such as 96, 93, and 20 etc, which are sparse data points and considered as anomalies.
- With -1 indicating an outlier, isolation forest detects 250 outliers，which is about 240 outliers more than the results from 3-sigma algorithms. Considering 3-sigma always rejects 0.3% of the input data set, the prediction made by isolation forest could be the result of unsupervised learning.
